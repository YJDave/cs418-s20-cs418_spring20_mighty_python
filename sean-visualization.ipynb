```python
data= data.fillna(0.0)
total_attendance = data["UGDS"]
median_debt =data["DEBT_MDN"].replace("PrivacySuppressed", 0.0)
data = data.drop('UGDS', axis =1)
data = data.drop('INSTNM', axis =1)
data = data.drop('DEBT_MDN', axis =1)
data = data.drop('DEBT_N', axis =1)
df_total_attendance = pd.DataFrame()
df_total_debt = pd.DataFrame()

for index, row in data.iterrows():
   df_total_attendance = df_total_attendance.append(row.multiply(total_attendance[index]))
   df_total_debt = df_total_debt.append(df_total_attendance.loc[index].multiply(float(median_debt[index])))

average = pd.DataFrame()
average =df_total_debt.sum(axis=0) / df_total_attendance.sum(axis=0)
#average.to_csv('Average debt by major', sep='\t')
```
